#### Generate some random data

```julia
make = rand(["Audi", "BMW", "Merc"], 100)
model = rand(["xz", "yz", "wz", "pz"], 100);
sales = rand(20:100, 100);
rating = rand(1:5, 100);
# data = DataFrame(Make = make, Model = model, Sales = sales, Rating = rating)
```

#### Create a line plot from the data

```julia
plot(make, sales, labels="Line plot")
```

#### Draw a scatter plot

```julia
scatter(make, sales, labels="Scatter Plot")
```

#### Draw a bar plot

```julia
plot(make, sales, seriestype=:bar)
```

#### Draw one plot above another

```julia
plot!(make, sales, seriestype=:line)
```

```julia
plot!(make, sales, seriestype=:scatter)
```

#### Draw multiple plots from some good data

```julia
x = ["a", "b", "c", "d", "e",]
y = rand(5,3)
```

```julia
plot(x,y[:,1], seriestype=:bar)
```

```julia
plot!(x,y[:,2], seriestype=:line)
```

```julia
plot!(x,y[:,3], seriestype=:scatter)
```

```julia
plot!(x,y[:,3], seriestype=:scatter, maker=:diamond)
```

#### Add labels to Plots

```julia
plot!(x,y[:,3], seriestype=:scatter, title="My Plotting", xlabel = "Alphabets", ylabel = "Numbers")
```

```julia
plot(cos, -4pi, 4pi, label="Cosine Function")
```

#### Draw multiple plots on a single frame

```julia
p1 = plot(x,y[:,1], seriestype=:bar)
p2 = plot(x,y[:,2], seriestype=:scatter)
p3 = plot(x,y[:,2], seriestype=:line)
p4 = plot(x,y[:,2], seriestype=:scatter, maker=:diamond)
plot(p1, p2, p3, p4)
```
